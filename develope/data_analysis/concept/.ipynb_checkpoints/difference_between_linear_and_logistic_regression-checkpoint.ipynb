## 선형 회귀 vs 로지스틱 회귀 비교

| 구분 | 선형 회귀 (Linear Regression) | 로지스틱 회귀 (Logistic Regression) |
|------|-----------------------------|----------------------------------|
| **출력값** | 연속형 값 (실수, 예: 혈당 수치 120mg/dL) | 범주형 값 (0/1, 예: 정상/고혈당) |
| **수학적 모형** | y = β0 + β1x1 + β2x2 + ... | P(y=1) = 1 / (1 + e^-(β0 + β1x1 + ...)) |
| **해석** | 입력값이 1 단위 증가할 때, 결과값이 얼마나 변하는지 | 입력값이 1 단위 증가할 때, 결과가 1(참)일 확률 변화 |
| **용도** | 예측(회귀): 혈당 수치, 체중, 칼로리 소비량 등 | 분류(Classification): 정상/고혈당, 질병 유무, 합격/불합격 등 |
| **손실 함수** | MSE (Mean Squared Error, 평균제곱오차) | Log Loss (이진 크로스 엔트로피) |
| **결과 해석** | 예측값 자체를 그대로 사용 | 예측 확률을 임계값(예: 0.5) 기준으로 분류 |
| **적합한 경우** | “얼마나?”를 알고 싶을 때 (정량 예측) | “맞다/아니다?”를 알고 싶을 때 (이진 분류) |
