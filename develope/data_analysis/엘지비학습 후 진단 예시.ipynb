```python
import pandas as pd
import pickle

# -----------------------------
# 모델 불러오기
# -----------------------------
with open("glucose_model.pkl", "rb") as f:
    model, meta = pickle.load(f)

# -----------------------------
# 사용자 입력 진단 함수
# -----------------------------
def predict_glucose(user_input: dict, threshold=0.4):
    """
    user_input: dict
        예: {"성별코드":1, "연령대코드(5세단위)":9, "신장":170, "체중":65,
             "수축기혈압":120, "이완기혈압":80}
    """
    # 1. 중앙값 기반 기본 데이터 생성
    input_data = meta["median_values"].copy()

    # 2. 사용자 입력 덮어쓰기
    for k, v in user_input.items():
        if k in input_data:
            input_data[k] = v

    # 3. 파생변수 생성
    if "신장" in user_input and "체중" in user_input:
        input_data["BMI"] = user_input["체중"] / ((user_input["신장"]/100) ** 2)
    if "수축기혈압" in user_input and "이완기혈압" in user_input:
        input_data["혈압차"] = user_input["수축기혈압"] - user_input["이완기혈압"]

    # 4. DataFrame 변환 (피처 순서 맞추기)
    df_input = pd.DataFrame([input_data])
    df_input = df_input.reindex(columns=meta["features"], fill_value=0)

    # 5. 예측
    proba = model.predict_proba(df_input)[:,1][0]

    # 6. 등급 판정
    if proba < 0.3:
        risk_level = "주의 ⚠️"
    elif proba < 0.6:
        risk_level = "경고 🚨"
    else:
        risk_level = "위험 🔴"

    return {
        "probability": round(proba, 3),
        "risk_level": risk_level
    }
```

```python
user_example = {
    "성별코드": 1,
    "연령대코드(5세단위)": 9,
    "신장": 172,
    "체중": 70,
    "수축기혈압": 125,
    "이완기혈압": 85,
    "흡연상태": 1,
    "음주여부": 0
}

result = predict_glucose(user_example, threshold=0.4)
print(result)
# 출력 예시: {'probability': 0.643, 'risk_level': '위험 🔴'}
```
